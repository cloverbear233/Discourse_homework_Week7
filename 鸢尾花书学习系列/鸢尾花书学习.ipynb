### 结合以上做一个计算实验

```python
import statistics
def calc_stats(operation:str,*args:list,**kwargs:dict)->float:
    result=0
    #标准差计算
    if operation=='stdev':
        #总体
        if 'TYPE'in kwargs and kwargs["TYPE"]=='population':
            result=statistics.pstdev(args)
        #样本
        elif 'TYPE' in kwargs and kwargs["TYPE"]=='sample':
            result=statistics.stdev(args)
        else:
            raise ValueError('TYPE,either population or sample')
    #计算方差
    elif operation=='var':
        #总体
        if 'TYPE'in kwargs and kwargs["TYPE"]=='population':
            result=statistics.pvariance(args)
        #样本
        elif 'TYPE' in kwargs and kwargs["TYPE"]=='sample':
            result=statistics.variance(args)
        else:
            raise ValueError('TYPE,either population or sample')
    else:
        print('Unsupported operation')
        return None
    #保留小数位
    if "ROUND" in kwargs:
        result=round(result,kwargs['ROUND'])
    return result
# 计算总体标准差
print(calc_stats("stdev", 1, 2, 3, 4, 5, 6,TYPE = 'population', ROUND = 3))
# 计算样本标准差
print(calc_stats("stdev", 1, 2, 3, 4, 5, 6, TYPE = 'sample'))
# 计算总体方差
print(calc_stats("var", 1, 2, 3, 4, 5, 6,TYPE = 'population', ROUND = 4))
# 计算样本方差
print(calc_stats("var", 1, 2, 3, 4, 5, 6, TYPE = 'sample'))
```

## PYthon面向对象编程（OOP）

```python
#面向对象编程 (Object-Oriented Programming, OOP) 是一种编程范式，它将数据和操作数据的方法组合在一起，形成一个对象。
#打个比方，OOP 中的类 (class) 就好比成套餐具，相当一种模板。
# 盘子好比属性(attribute)用来装各种食物 (数据)；刀叉好比方法 (method)，用来用餐 (操作)。
class Rectangle:#定义矩阵类，有width,height两个属性，有circumference,area两个方法；用class把属性和方法封装
    def __init__(self,width,height):#初始化操作，在__init__方法的参数列表中，第一个参数通常被命名为 self，它指向类的实例对象。
        self.width=width
        self.height=height#self 参数在调用类的其他方法时自动传递，可以通过 self 访问类的属性和其他方法。
        #在__init__方法内部，可以定义初始化对象时需要执行的逻辑，例如设置对象的初始状态，为对象设置属性的初始值等。
    def circumference(self):
        return 2*(self.width+self.height)
    def area(self):
        return self.width*self.height
#创建rect1作为对象实例
rect1=Rectangle(5,10)
print('矩形周长:',end='')
print(rect1.circumference())
print('矩形面积:',end='')
print(rect1.area())
```

## 定义方法

```python
#定义一个 ListStatistics 类来计算一个浮点数列表的长度、和、平均值、方差。
class ListStatistics:
    def __init__(self,data):
        # ListStatistics包含一个data属性来存储浮点数列表
        self.data=data
    def list_length(self):
        return len(self.data)
    def list_sum(self):
        return sum(self.data)
    def list_mean(self):
        return sum(self.data)/self.list_length()
    def list_variance(self,ddof=1):
        sum_squares=sum((x_i-self.list_mean())**2 for x_i in self.data)
        return sum_squares/(self.list_length()-ddof)
#创建一个浮点数列表
data=[8.8,1.8,7.8,2.8,5.6,3.9,6.9,3.8]
#创建ListStatistics对象实例
float_list=ListStatistics(data)
# 使用float_list对象计算列表长度
print("列表长度：", float_list.list_length())
# 使用float_list对象计算列表和
print("列表和：", float_list.list_sum())
# 使用float_list对象计算列表平均值
print("列表平均值：", float_list.list_mean())
# 使用float_list对象计算列表方差
print("列表方差：", float_list.list_variance())
print("列表方差 (ddof = 0)：",float_list.list_variance(0))
```

```python
# 导入包
import numpy as np
import matplotlib.pyplot as plt
# 生成横轴数据（这里的numpy.linspace()函数用于生成一个等差数列，从0到2π，分成100个点）
x_array = np.linspace(0, 2*np.pi, 100)
# 正弦函数数据
sin_y = np.sin(x_array)
# 余弦函数数据
cos_y = np.cos(x_array)
# 设置图片大小
fig, ax = plt.subplots(figsize=(8, 6))
# 绘制正弦和余弦曲线
ax.plot(x_array, sin_y,
label='sin', color='b', linewidth=2)
ax.plot(x_array, cos_y,
label='cos', color='r', linewidth=2)
# 设置标题、横轴和纵轴标签
ax.set_title('Sine and cosine functions')
ax.set_xlabel('x')
ax.set_ylabel('f(x)')
# 添加图例
ax.legend()
# 设置横轴和纵轴范围
ax.set_xlim(0, 2*np.pi)
ax.set_ylim(-1.5, 1.5)
# 设置横轴标签和刻度标签
x_ticks = np.arange(0, 2*np.pi+np.pi/2, np.pi/2)
x_ticklabels = [r'$0$', r'$\frac{\pi}{2}$',
r'$\pi$', r'$\frac{3\pi}{2}$',
r'$2\pi$']
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_ticklabels)
# 横纵轴采用相同的scale
ax.set_aspect('equal')
plt.grid()
# 将图片存成SVG格式
plt.savefig('正弦_余弦函数曲线.svg', format='svg')
# 显示图形
plt.show()
```
