### **T1** -offer21、调整数组顺序使奇数位于偶数前面

```python
#1 简单法（略，直接找两个表一个排过去）
def exchange1(nums:list[int])->list[int]:
    nums0=[]
    nums1=[]
    for i in nums:
        if i%2==1:
            nums0.append(i)
        else:
            nums1.append(i)
    return nums0+nums1
#2 头尾双指针
def exchange2(nums:list[int])->list[int]:
    left,right=0,len(nums)-1
    while left<right:
        while nums[left]%2==1:
            left+=1
        while nums[right]%2==0:
            right-=1
        nums[left],nums[right]=nums[right],nums[left]#两边交换完在中间汇合
    return nums
#3 快慢指针
def exchange3(nums:list[int])->list[int]:
    fast,slow=0,0
    while fast<len(nums):
        if nums[fast]%2==1:
            nums[slow],nums[fast]=nums[fast],nums[slow]
            slow+=1#slow位置填充fast识别的奇数，然后进到下一格等着填充下一个识别
        fast+=1
    return nums
numstest=[16,2,5,3,6,8,7,1,9]
print(exchange1(numstest))
print(exchange2(numstest))
print(exchange3(numstest))
```

### **T2** 文件组合
待传输文件被切分成多个部分，按照原排列顺序，每部分文件编号均为一个 正整数（至少含有两个文件）。传输要求为：连续文件编号总和为接收方指定数字 target 的所有文件。请返回所有符合该要求的文件传输组合列表。

注意，返回时需遵循以下规则：

每种组合按照文件编号 升序 排列；
不同组合按照第一个文件编号 升序 排列。

```python
#法1 暴力法
def fileCombination1(target: int) -> list[list[int]]:
        listall=[]
        for i in range(1,target):
            list1=[i]
            sum=i
            for j in range(i+1,target):
                sum+=j
                if sum<target:
                    list1.append(j)
                    continue
                elif sum==target:
                    list1.append(j)
                    listall.append(list1)
                else:
                    break
        return listall
#法2 从公式出发改进(O(n2)和O(n))
def fileCombination2(target: int) -> list[list[int]]:
        listall=[]
        for i in range(1,target):
            for k in range(1,target-i):
                if (i+k+i)*(k+1)==target*2:
                     listall.append(list(range(i,k+i+1)))
                elif (i+k+i)*(k+1)>target*2:
                     break
        return listall
#法3 复杂度O(n),O(n)
'''
算法流程：
1\初始化： 左边界 i=1 ，右边界 j=2 ，元素和 s=3 ，结果列表 res ；
2\循环： 当 i≥j 时跳出；
 当 s>target 时： 向右移动左边界 i=i+1 ，并更新元素和 s ；
 当 s<target 时： 向右移动右边界 j=j+1 ，并更新元素和 s ；
 当 s=target 时： 记录连续整数序列，并向右移动左边界 i=i+1 ；
3\返回值： 返回结果列表 res ；
'''

def fileCombination3(target:int)->list[list[int]]:
    i,j,sum,listall=1,2,3,[]
    while i<j:
        if sum==target:
            listall.append(list(range(i,j+1)))
        if sum>=target:
            sum-=i
            i+=1
        else:
            j+=1
            sum+=j
    return listall

targettest=30250
print(fileCombination1(targettest))
print(fileCombination2(targettest))
print(fileCombination3(targettest))
```

### **T3** 按规则计算统计结果
为了深入了解这些生物群体的生态特征，你们进行了大量的实地观察和数据采集。数组 arrayA 记录了各个生物群体数量数据，其中 arrayA[i] 表示第 i 个生物群体的数量。请返回一个数组 arrayB，该数组为基于数组 arrayA 中的数据计算得出的结果，其中 arrayB[i] 表示将第 i 个生物群体的数量从总体中排除后的其他数量的乘积。

```python
#法一：一般讨论特殊情况的除法辨别法（复杂度O（n)，O（n））
def statisticalResult1(arrayA: list[int]) -> list[int]:
        all=1
        arrayB=[]
        for num in arrayA:
            all*=num
        if all==0:
            count=0
            com=1
            set=0
            for i in range(len(arrayA)):
                if arrayA[i]==0:
                    count+=1
                    set+=i
                else:
                    com*=arrayA[i]
                if count==1:
                    arrayB=[0]*len(arrayA)
                    arrayB[set]=com
                else:
                    arrayB=[0]*len(arrayA)
        else:
            for i in range(len(arrayA)):
                arrayB.append(int(all/arrayA[i]))
        return arrayB
#法2、不运用除法的算法设计

arraytest1=[2,4,6,8,10]
arraytest2=[0,2,4,8,10]
print(statisticalResult1(arraytest1))
print(statisticalResult1(arraytest2))
```
