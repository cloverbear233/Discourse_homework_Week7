```python
#1.反转字典
my_dict={
    "brand":"Ford",
    "model":"Mustang",
    "year":1964
}
print(dict(map(reversed,my_dict.items())))
```

```python
#2.对字典列表进行排序
dicts_list=[
    {"name":"Tom","age":20},
    {"name":"Nick","age":25},
    {"name":"John","age":30}
]
print(sorted(dicts_list,key=lambda x:x["age"]))
dicts_list.sort(key=lambda item:item.get("age"))
print(dicts_list)
```

```python
#3.根据另一个列表对列表进行排序
a=["blue","green","orange","purple","yellow"]
b=[3,2,5,4,1]
print(list(zip(b,a)))
```
